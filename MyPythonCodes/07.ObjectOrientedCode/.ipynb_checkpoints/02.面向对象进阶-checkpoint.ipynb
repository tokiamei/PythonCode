````python
import math

class Triangle(object):
	def __init__(self, a, b, c):
		self._a = a
		self._b = b
		self._c = c
	
	@staticmethod
	def is_valid(a, b, c):
		return a + b > c and b + c > a and a + c > b
	def perimeter(self):
		return self._a + self._b + self._c
	def area(self):
		half = self.perimeter() / 2
		return sqrt(half * (half - self._a) * (half - self._b) * (half - self._c))

a, b, c = 3, 4, 5
if Triangle.is_valid(a, b, c):
	t = Triangle(a, b, c)
	print(t.perimeter())
	print('+++', t.is_valid(a, b, c))
else:
	print('æ— æ³•æ„æˆä¸‰è§’å½¢')
````

# ç±»æ–¹æ³•

- æœ‰çš„æ–¹æ³•ä¸å±äºæŸä¸ªå¯¹è±¡ï¼Œå±äºç±»

````python
from time import time, localtime, sleep

class Clock(object):
	"""æ•°å­—æ—¶é’Ÿ"""
	def __init__(self, hour = 0, minute = 0, second = 0):
		self._hour = hour
		self._minute = minute
		self._second = second
	# ç¬¬ä¸€ä¸ªå‚æ•°çº¦å®šåä¸º clsï¼Œå®ƒä»£è¡¨å½“å‰ç±»ç›¸å…³çš„ä¿¡æ¯çš„å¯¹è±¡ã€ç±»æœ¬èº«ä¹Ÿæ˜¯ä¸€ä¸ªå¯¹è±¡ï¼Œæœ‰çš„åœ°æ–¹ç§°ä¹‹ä¸ºç±»çš„å…ƒæ•°æ®å¯¹è±¡ã€‘ï¼Œé€šè¿‡å‚æ•°æˆ‘ä»¬å‘¢å¯ä»¥è·å–å’Œç±»ç›¸å…³çš„ä¿¡æ¯å¹¶ä¸”å¯ä»¥åˆ›é€ å‡ºç±»çš„å¯¹è±¡
	
	@classmethod
	def now(cls):
		ctime = localtime(time())
		return cls(ctime.tm_hour, ctime.tm_min, ctime.tm_sec)
	
	def run(self):
		"""èµ°å­—"""
		self._second += 1
		if self._second == 60:
			self._second = 0
			self._minute += 1
			if self._minute == 60:
				self._minute = 0
				self._hour += 1
				if self._hour == 24:
					self._hour = 0
	def show(self):
		"""æ˜¾ç¤ºæ—¶é—´"""
		return '%02d:%02d:%02d' % (self._hour, self._minute, self._second)

# é€šè¿‡ç±»æ–¹æ³•åˆ›å»ºå¯¹è±¡å¹¶è·å–ç³»ç»Ÿæ—¶é—´
clock = Clock.now()
while True:
	print(clock.show())
	sleep(1)
	clock.run()
````

# ç»§æ‰¿

- åœ¨å·²æœ‰ç±»çš„åŸºç¡€ä¸Šåˆ›å»ºæ–°ç±»ï¼Œè®©ä¸€ä¸ªç±»ä»å¦ä¸€ä¸ªç±»é‚£é‡Œå°†å±æ€§å’Œæ–¹æ³•ç›´æ¥ç»§æ‰¿ä¸‹æ¥
- æœ‰æ•ˆçš„å‡å°å¼€å‘çš„å·¥ä½œé‡ï¼Œåˆ©äºä»£ç çš„ç»´æŠ¤

````python
class Person(object):
	"""äºº"""
	def __init__(self, name, age):
		self._name = name
		self._age = age
	@property
	def name(self):
		return self._name
	@property
	def age(self):
		return self._age
	@age.setter
	def age(self, age):
		self._age = age
	def play(self):
		print('%sæ­£åœ¨æ„‰å¿«çš„ç©è€' % self._name)
	def watch_tv(self):
		if self._age >= 18: print('%sæ­£åœ¨çœ‹é’¢é“ä¾ ' % self._name)
		else: print('%sæ­£åœ¨çœ‹çŒªçŒªä¾ ' % self._name)

class Student(Person):
	"""å­¦ç”Ÿ"""
	def __init__(self, name, age, grade):
		super().__init__(name, age)
		self._grade = grade
	@property
	def grade(self):
		return self._grade
	@grade.setter
	def grade(self, grade):
		self._grade = grade
	def study(self, course):
		print('%så¹´çº§çš„%sæ­£åœ¨å­¦ä¹ %s' % (self._grade, self._name, course))
		
class Teacher(Person):
	"""è€å¸ˆ"""
	def __init__(self, name, age, title):
		super().__init__(name, age)
		self._title = title
	@property
	def title(self):
		return self._title
	@title.setter
	def title(self, title):
		self._title = title
	def teach(self, course):
		print('%sæ­£åœ¨è®²æˆ%s' % (self._name, course))

stu = Student('HHH', 5, 3)
stu.watch_tv()
````

- å¯¹çˆ¶ç±»å·²æœ‰çš„æ–¹æ³•è¿›è¡Œé‡å†™ã€ç»™å‡ºæ–°çš„å®ç°ç‰ˆæœ¬ã€‘
- å½“æˆ‘ä»¬è°ƒç”¨è¿™ä¸ªç»è¿‡å­ç±»é‡å†™çš„æ–¹æ³•æ—¶ï¼Œä¸åŒå­ç±»å¯¹è±¡ä¼šè¡¨ç°å‡ºä¸åŒçš„è¡Œä¸ºï¼Œè¿™ä¸ªå°±æ˜¯å¤šæ€ã€poly-morphismã€‘



````python
from abc import ABCMeta, abstractmethod

# å°† Pet å°è£…ä¸ºæŠ½è±¡ç±»ã€æŠ½è±¡ç±»ä¸èƒ½å®ä¾‹åŒ–ï¼ŒæŠ½è±¡ç±»çš„å­˜åœ¨å°±æ˜¯ä¸ºäº†å…¶ä»–ç±»å»ç»§æ‰¿å®ƒã€‘
class Pet(object, metaclass = ABCMeta):
	"""å® ç‰©"""
	def __init__(self, nickname):
		self._nickname = nickname
	@property
	def nickname(self):
		return self._nickname
	@abstractmethod
	def make_voice(self):
		"""å‘å‡ºå£°éŸ³"""
		pass

class Dog(Pet):
	"""ç‹—"""
	def make_voice(self):
		print('%s: æ±ªæ±ªæ±ª' % self._nickname)

class Cat(Pet):
	"""çŒ«"""
	def make_voice(self):
		print('%s: å–µå–µå–µ' % self._nickname)

pets = [Dog('æ—ºè´¢'), Cat('å‡¯è’‚'), Dog('å¤§é»„')]
for pet in pets:
	pet.make_voice()
````
