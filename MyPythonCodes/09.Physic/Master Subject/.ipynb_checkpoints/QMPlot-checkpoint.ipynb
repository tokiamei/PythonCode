```python
import os
import re
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use('science')
```

```python
# 导入数据
def getData(file):
    with open(file, 'rt') as raw_data:
    	data = np.loadtxt(raw_data, delimiter=',')
    return data
```

```python
# 取出文件名当作变量名并对其赋值
dirList = os.listdir()
for item in dirList:
    if re.search(r'.*\.DAT', item): locals()[item.rsplit('.')[0]] = getData(item)
```

```python
position = np.linspace(-100, 100, 513)
```

```python
def DensityPlot(DensityUp, DensityDown, AtomNumber, Density, Factor = 1.3):
    DensityUpMax = np.max(DensityUp)
    DensityDownMax = np.max(DensityDown)
    DensityUpMin = np.min(DensityUp)
    DensityDownMin = np.min(DensityDown)
    # print(max(DensityUpMax, DensityDownMax), min(DensityUpMin, DensityDownMin))
    with plt.style.context(["science", "ieee", "no-latex"]):

        plt.plot(position, DensityUp, label=r'$\phi_{' + AtomNumber + '\\uparrow}$')    
        plt.plot(position, DensityDown, label=r'$\phi_{' + AtomNumber + '\\downarrow}$')    
        plt.legend(loc='upper right')
        plt.xlabel('position')
        plt.ylabel(Density + ' Distribution')
        plt.ylim(min(DensityUpMin, DensityDownMin)*1.1, max(DensityUpMax, DensityDownMax)*Factor)
        plt.title(Density + ' Function Of Condensate ' + AtomNumber)
        plt.show()

DensityPlot(DENA1, DENA2, "A", 'Density')
DensityPlot(DENB1, DENB2, "B", 'Density')
DensityPlot(PHASEB1, PHASEB2, "B", 'Phase', 1.6)
```
