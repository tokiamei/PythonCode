```python
from multiprocessing import Pool
# import multiprocess
import os, time
```

```python
def task(id):
    # print(f'子进程{os.getpid}工作在任务{id}上')
    time.sleep(2)
    
if __name__ == '__main__':
    print('程序开始！')
    p = Pool() # create process pool
    for i in range(20):
        p.apply_async(task, args=(i,))
    p.close() # close process pool
    p.join()
    print('程序执行完毕！')
```
