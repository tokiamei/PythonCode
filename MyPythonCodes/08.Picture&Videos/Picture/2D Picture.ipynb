# 一、折线图

## 1、简单绘制折线图

```python
x = range(1, 7)
y = [13, 15, 14, 16, 15, 17]

'''
    figsize：设置图片的宽、高，单位为英寸
    dpi：设置分辨率
'''
plt.figure(figsize=(4, 2), dpi=500)
# plt.title('折线图')
plt.xlabel('x Axis')
plt.ylabel('y Axis')
plt.xlabel

'''
    color：颜色
    linewidth：线的宽度
    marker：折点样式
    linestyle：线的样式，主要包括：'-'、'--'、'-.'、':'
'''
plt.plot(x, y, color='#f6a5c0', marker='*', linewidth='1', linestyle='--')
# 保存
# plt.savefig('test.png')
plt.show()
```

## 2、多线图

```python
x = range(15, 25)
y1 = [50, 55, 58, 65, 70, 68, 70, 72, 75, 70]
y2 = [52, 53, 60, 63, 65, 68, 75, 80, 85, 72]

plt.figure(figsize=(8, 5), dpi=500, facecolor='lightgoldenrodyellow')

plt.plot(x, y1, color='red', label='Tom')
plt.plot(x, y2, color='blue', label='Jerry')

# plt.title('体重年龄折线图')
plt.xlabel('Age(Years Old)')
plt.ylabel('Weight(Kg)')
plt.grid(alpha=0.2) # 添加网格，alpha为透明度
plt.legend(loc='upper left') # 添加图例
plt.show()
```

## 3、绘制子图

```python
a = np.arange(1, 30)
# 划分子图
fig, axs = plt.subplots(2, 2, figsize = (8, 5), dpi = 500)


# 绘制子图
axs[0, 0].plot(a, a, label = "line")
axs[0, 1].plot(a, np.sin(a), label = 'sine')
axs[1, 0].plot(a, np.log(a), label = 'log')
axs[1, 1].plot(a, a ** 2, label = 'power')
# plt.legend()
plt.figure().legend()
```

## 2、多变量散点图

```python
import scienceplots
plt.style.use('science')

num_points = 100 # 设置散点的数量

# 使用NumPy生成随机分布的x, y坐标以及对应的z值
x = np.random.rand(num_points)
y = np.random.rand(num_points)
z = y
# z = np.random.rand(num_points)  # 假设这是影响颜色的第三个变量

with plt.style.context(["science", "ieee", "no-latex"]):
    # 创建一个颜色映射（ colormap ），这里使用 'viridis' ，也可以选择其他colormap
    cmap = plt.cm.get_cmap('viridis')

    colors = cmap(z) # 使用z值作为颜色数据，范围通常会归一化到0-1之间
    fig = plt.figure(figsize=(8, 5), dpi=600)
    # ax = fig.subplots()
    # s=30 指定了散点的大小为 30 像素单位。
    # c=colors表示散点的颜色由变量colors决定，这个变量应该是一个和数据点一一对应的颜色列表或者数组，其值对应于z变量经过色彩映射后的颜色值。
    # alpha=0.3 设置了散点的透明度，值越小越透明。
    #cmap=cmap 这里指定了用于颜色映射的色彩方案
    plt.scatter(x, y, s=30, c=colors, alpha=0.3, cmap=cmap)
    # 添加一个色条（colorbar）显示颜色与z值的关系
    plt.colorbar(label='Z-value')

    # plt.title('Randomly Distributed Scatter Plot with Color Mapping')
    # plt.xlabel('X-axis')
    # plt.ylabel('Y-axis')
    # plt.show()
```

## 3、一张图内多条线与散点图

```python
x = np.linspace(0, 10, 100)
y1 = x**2
y2 = x + np.sin(x)
y3_scatter = np.random.rand(50) * 5
x_scatter = np.random.rand(50) * 10

'''
创建一个新的图形，通过 plt.subplots() 函数创建了一个新的 figure，并且在这个 figure 中设置了一个单个的 Axes 对象，
并将其引用赋值给变量 ax。
这样，通过 ax 我们就可以方便地在这个单一的绘图区（axes）上进行各种操作，如绘制曲线、添加图例、调整坐标轴范围等等
'''
fig, ax = plt.subplots(figure=(8, 5), dpi=500)

ax.plot(x, y1, label='Curve 1', color='blue')  # 第一条曲线
ax.plot(x, y2, label='Curve 2', color='red')   # 第二条曲线
ax.scatter(x_scatter, y3_scatter, color='green', marker='o', label='Scatter Points')

ax.legend()
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Multiple Lines and Scatter Plot on the Same Figure')

# 显示图形
plt.show()
```

# 三、坐标轴

```python
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(axes_class=AxesZero)

for direction in ["xzero", "yzero"]:
    # adds arrows at the ends of each axis
    ax.axis[direction].set_axisline_style("-|>")

    # adds X and Y-axis from the origin
    ax.axis[direction].set_visible(True)

for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax.axis[direction].set_visible(True)

x = np.linspace(-0.5, 1., 100)
ax.plot(x, np.sin(x*np.pi))

plt.show()
```
