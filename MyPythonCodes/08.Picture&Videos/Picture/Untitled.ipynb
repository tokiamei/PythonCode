```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph.
    """
    out = ax.plot(data1, data2, **param_dict)
    # return out


data1, data2, data3, data4 = np.random.randn(4, 100)  # make 4 random data sets
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'});
```

```python
dataDict :dict = {
    'P1': np.random.rand(15)*10 + 20,
    'P2': np.random.rand(15)*10 + 30.55,
    'P3': np.random.rand(15)*10 + 40,
    'P4': np.random.rand(15)*10 + 50,
    'P5': np.random.rand(15)*10 + 40,
    'P6': np.random.rand(15)*10 + 30,
    'P7': np.random.rand(15)*10 + 20,
    'P8': np.random.rand(15)*10 + 40,
    'P9': np.random.rand(15)*10 + 50,
    'P10': np.random.rand(15)*10 + 30
}
labels :list = list(dataDict.keys())
data :list = list(dataDict.values())

wafNumbers = len(labels)
length = len(data[0])

averageList :list = [np.average(iterm) for iterm in data]

fig, ax = plt.subplots(figsize=(9, 4), dpi=600)

ax.plot(range(1, wafNumbers + 1), averageList, '--')

ax.boxplot(
    x=data,
    vert=True,
    widths=0.2,
    tick_labels=labels,
    patch_artist=False,
    medianprops={'linestyle': '-', 'color': 'r', 'linewidth': 1.5}, 
    showmeans=True,
    # meanline=True,
    meanprops={'marker': 'o', 'markersize': 5, 'markeredgewidth': 0.75, 'markerfacecolor': '#b7e1a1', 'markeredgecolor': 'r', 'color': 'k', 'linewidth': 1.5},
    showfliers=True,
    flierprops={'marker': '+', 'markersize': 6.75, 'markeredgewidth': 0.75, 'markerfacecolor': '#ee5500', 'markeredgecolor': 'k'}, 
    whiskerprops={'linestyle': '--', 'linewidth': 1, 'color': '#480656'},
    capprops={'linestyle': '-', 'linewidth': 1.5, 'color': '#480656'}
)


list(ax.scatter(np.ones(length) * i - 0.2, data[i-1], marker='^', linewidths=0.1) for i in range(1, wafNumbers + 1))

ax.yaxis.grid(True)
ax.set_xlabel("WafName")
ax.set_ylabel(r"$Dit\ \ C-\phi$")
```
