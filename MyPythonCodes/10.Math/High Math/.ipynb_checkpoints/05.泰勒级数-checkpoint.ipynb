# 1、泰勒级数

    计算泰勒级数前 20 项的和

```python
# 用指数函数演示
from sympy.abc import x 
import numpy as np
exp = np.e**x
sum = 0
for i in range(20):
    derivative = exp.diff(x, i)
    derivative = derivative.evalf(subs={x: 0})
    facto = np.math.factorial(i)
    sum += derivative/facto * x**i
# sum

# # np.e**x.evalf(subs={x:1})
# 验证是否正确 首先 x = 0 的情况
print(sum.evalf(subs={x: 0}) - exp.evalf(subs={x: 0}))
```

```python
# 作图比较
import matplotlib.pyplot as plt 
xVals = np.linspace(0, 20, 100)
plt.figure(figsize=(8, 5), dpi = 500)
for xVal in xVals:
    plt.plot(xVal, sum.subs(x, xVal), 'o', \
             xVal, exp.subs(x, xVal), '+')
```

# 研究和项数的关系

```python
from sympy.abc import x 
import numpy as np
exp = np.e**x

def expTaylor(count):
    sum = 0
    for i in range(count):
        derivative = exp.diff(x, i)
        derivative = derivative.evalf(subs={x: 0})
        facto = np.math.factorial(i)
        sum += derivative/facto * x**i
    return sum
expTaylor(20).subs(x, 1)

import matplotlib.pyplot as plt
xVals = np.linspace(0, 20, 100)
plt.figure(figsize=(8, 5), dpi = 500)
plt
for xVal in xVals:
    plt.plot(xVal, expTaylor(10).subs(x, xVal), '+', \
             xVal, expTaylor(15).subs(x, xVal), 'o', \
             xVal, expTaylor(20).subs(x, xVal), '-', \
             xVal, exp.subs(x, xVal))

```
