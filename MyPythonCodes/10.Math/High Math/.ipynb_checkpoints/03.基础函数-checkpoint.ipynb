# 1、多项式的简单小例子

$$f(x) = x^3 - 5x^2 + 9$$

```python
# 多项式
def f(x):
    return x**3 -5*x**2 + 9
f(1)
```

```python
# 作图
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5, 5, 100)
y = f(x)
plt.figure(figsize=(8, 5), dpi = 500)
plt.plot(x, y)
plt.show()
```

# 2、指数函数

$$f(x) = e^x$$

```python
# 函数方式进行创建
import numpy as np
# def g(x):
#     return (np.e)**x
```

```python
# 用 np 自带的指数函数
def g(x):
    return np.exp(x)
    
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (8, 5), dpi = 500)
# x = np.linspace(-5, 5, 100)
x = np.arange(-5, 5, 0.1)

y = g(x)
plt.plot(x, y)
plt.show()
```

# 3、e 指数的表示

    用泰勒级数来表示

```python
# import numpy as np
def exp(x):
    sum = 0
    for k in range(1000):
        sum = sum + x**k/(np.math.factorial(k))
    return sum
exp(1), np.e # 精确度还是很高的
```

# 4、对数函数

```python
import numpy as np
# 忽略除零警告
np.seterr(divide = 'ignore') 
x = np.linspace(0, 10, 100, endpoint = False)
y1 = np.log2(x)
y2 = np.log(x)
y3 = np.log10(x)
import matplotlib.pyplot as plt
plt.figure(figsize = (8, 5), dpi = 500)
plt.plot(x, y1, x, y2, x, y3)
plt.show()
```
