```python
import numpy as np, matplotlib.pyplot as plt, scienceplots, sympy as sp
# 使用Science风格的配置文件
plt.style.use('science')
```

# 一、初等函数绘图

```python
# Create Function
x = np.array(np.arange(-5, 5.2, 0.2))
y = x**2
```

```python
def functionPlot(xAxis, yAxis, plotLabel = "Function Curve", xLabel = "x", yLabel = "y", title = "Function Curve"):
    with plt.style.context(["science", "ieee", "no-latex"]):
        plt.plot(xAxis, yAxis, label = rf'{plotLabel}')
        plt.legend(loc='upper right')
        plt.xlabel(xLabel)
        plt.ylabel(yLabel)
        plt.title(title)
        plt.show()
```

# 二、极限

```python
x = sp.symbols('x')
f1 = sp.sin(x)/x
L = sp.limit(f1, x, 'oo')
print(f'{f1}的极限是{L}')

x1 = np.arange(-100, 100, 0.01)
y1 = np.sin(x1)/x1
functionPlot(x1, y1, "$y1 = sin(x1)/x1$", "x1", "y1")
```

# 三、求导

## 1、按照导数定义求解

```python
def numerical_diff(f, x):
    h = 1e-4
    return (f(x + h) - f(x - h))/(2*h)

def func_1(x):
    return x**3

d = numerical_diff(func_1, 5)

print(d)
```

## 2、Sympy内置求导函数求解

```python
x = sp.symbols('x')
expr = x**2
# expr = sp.sin(x)
d = sp.diff(expr, x)
print(d)
print(d.evalf(subs={'x':5}))
```

## 3、多阶导数

```python
x = sp.symbols('x')
# expr = sp.sin(x)/x
expr = x**2
d1 = sp.diff(expr, x)
print(d1)
d2 = sp.diff(expr, x, 2) # 2'd
print(d2)
d3 = sp.diff(expr, x, 3) # 2'd
print(d3)
```

## 4、画切线

```python
# original line
def f(x):
    return x**2

# tengent line
def tangent_line(f, x0):
    a = numerical_diff(f, x0)
    b = f(x0) - a*x0
    def line(t):
        return a*t + b
    return line

x = np.arange(-20, 20, 0.01)
y1 = f(x)
y2 = tangent_line(f, 5)(x)

with plt.style.context(["science", "ieee", "no-latex"]):
        plt.plot(x, y1, label = r'y1')
        plt.plot(x, y2, label = r'y2')
        plt.legend(loc='upper right')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Origin And Tangent Lines')
        plt.show()
```

## 5、隐函数求导

```python
x, y = sp.symbols('x y')
z = x**2 + y**2
f = sp.idiff(z, y, x) # y对x求导
print(f)
```
